```python
# escrever um programa que o usuario digita uma palavra e o programa verifica se
# a palavra é palindroma e imprime a palavra.

# Entrada

palavra = str (input("digite uma palavra: "))
invertida = ""

for i in (palavra):

  invertida = i + invertida

if invertida == palavra:

  print (f" {invertida} é uma palavra palindroma ")

else:

  print (f" {palavra} não é uma palavra palindroma ")
```
