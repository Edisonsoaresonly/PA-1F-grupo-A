```python
# classificação de idade

# Escrever um programa que recebe o nome e a idade do usuário e verifica se o usuário é menor de idade, adulto ou idoso

# <= 17 anos é menor de idade
# >= 18 anos é adulto
# >= 60 anos é idoso

# Entrada

nome = str(input("digite o seu nome: "))
idade = int(input("digite sua idade: "))

# Processamento / Saída

if idade <= 17:
 print(f"{nome}, você tem {idade} anos de idade, você é menor de idade.")

elif idade >= 60:
  print(f"{nome}, você tem {idade} anos de idade, você é idoso.")

else:
  print(f"{nome}, você tem {idade} anos de idade, você é adulto.")
```

```python
# Cálculo de desconto

# Entrada do valor da compra
valor_compra = float(input("Digite o valor da compra: R$"))

# Variável para armazenar o valor final
valor_final = 0

# Aplicação das regras de desconto
if valor_compra <= 100:
    valor_final = valor_compra
    print(f"Valor da compra: R${valor_compra:.2f}")
    print("Não há desconto aplicado.")
elif valor_compra <= 500: # Isso cobre de R$100.01 até R$500.00
    desconto = valor_compra * 0.10
    valor_final = valor_compra - desconto
    print(f"Valor da compra: R${valor_compra:.2f}")
    print(f"Desconto de 10% (R${desconto:.2f}) aplicado.")
elif valor_compra > 500:
    desconto = valor_compra * 0.20
    valor_final = valor_compra - desconto
    print(f"Valor da compra: R${valor_compra:.2f}")
    print(f"Desconto de 20% (R${desconto:.2f}) aplicado.")

# Exibição do valor final
print(f"Valor final da compra: R${valor_final:.2f}")
```

```python
# Verificação de triângulo

# Entrada dos três valores
lado1 = float(input("Digite o valor do primeiro lado: "))
lado2 = float(input("Digite o valor do segundo lado: "))
lado3 = float(input("Digite o valor do terceiro lado: "))

# Verificação da condição do triângulo
if (lado1 + lado2 > lado3) and (lado1 + lado3 > lado2) and (lado2 + lado3 > lado1):
    print("Os valores podem formar um triângulo válido.")
else:
    print("Os valores NÃO podem formar um triângulo válido.")
```

```python
# Sistema de energia

# Entrada da energia do personagem
energia = int(input("Digite a energia do personagem (0 a 100): "))

# Classificação da energia
if energia == 0:
    print("Classificação: Morto")
elif 1 <= energia <= 30:
    print("Classificação: Crítico")
elif 31 <= energia <= 70:
    print("Classificação: Normal")
elif 71 <= energia <= 100:
    print("Classificação: Saudável")
else:
    print("Valor de energia inválido! Por favor, digite um número entre 0 e 100.")
```

```python
# Média escolar

# Entrada da média final
media = float(input("Digite a média final do aluno: "))

# Classificação da média
if media >= 7:
    print("Classificação: Aprovado")
elif media >= 5 and media < 7:
    print("Classificação: Recuperação")
else:
    print("Classificação: Reprovado")
```

```python
# Cálculo de salário com bônus

# Entrada do salário e pontuação de desempenho
salario = float(input("Digite o salário do funcionário: R$"))
pontuacao = int(input("Digite a pontuação de desempenho (0 a 10): "))

# Inicializa o valor do bônus
valor_bonus = 0

# Aplicação das regras de bônus
if pontuacao < 5:
    valor_bonus = 0  # Sem bônus
elif pontuacao >= 5 and pontuacao <= 7:
    valor_bonus = salario * 0.05  # 5% de bônus
elif pontuacao >= 8 and pontuacao <= 9:
    valor_bonus = salario * 0.10  # 10% de bônus
elif pontuacao == 10:
    valor_bonus = salario * 0.15  # 15% de bônus
else:
    print("Pontuação inválida! Por favor, digite um número entre 0 e 10.")
    valor_bonus = 0 # Define bônus como 0 para pontuações inválidas

# Cálculo do salário final
salario_final = salario + valor_bonus

# Exibição dos resultados
print(f"Valor do bônus: R${valor_bonus:.2f}")
print(f"Salário final com bônus: R${salario_final:.2f}")
```
