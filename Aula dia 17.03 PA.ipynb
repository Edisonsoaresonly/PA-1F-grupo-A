```python
#  Operadores aritiméticos
# Fazer um programa que calcula operações com 02 números

# Entrada

a = int (input("digite um número: "))
b = int (input("digite um número: "))

# Processamento

soma = a + b
subtracao = a - b
multiplicacao = a * b
divisao = a / b
div_int = a // b
resto = a % b
potencia = a ** b

# Saída
print ("")
print("-----------------------OPERADORES ARITIMÉTICOS-------------------------")
print("")
print("--------------------------------SOMA-----------------------------------")
print("")
print(f"O resultado da SOMA de {a} MAIS {b} é igual a : {soma}")
print("")
print("------------------------------SUBTRAÇÃO--------------------------------")
print("")
print(f"O resultado da SUBTRAÇÃO de {a} MENOS {b} é igual a : {subtracao}")
print("")
print("------------------------------MULTIPLICAÇÃO----------------------------")
print("")
print(f"O resultado da MULTIPLICAÇÃO de {a} VEZES {b} é igual a : {multiplicacao}")
print("")
print("-------------------------------DIVISÃO---------------------------------")
print("")
print(f"O resultado da DIVISÃO de {a} DIVIDIDO {b} é igual a : {divisao}")
print("")
print("----------------------------DIVISÃO INTEIRA----------------------------")
print("")
print(f"O resultado da DIVISÃO INTEIRA de {a} DIVIDIDO {b} é igual a : {div_int}")
print("")
print("----------------------------RESTO DA DIVISÃO---------------------------")
print("")
print(f"O resultado da DIVISÃO INTEIRA de {a} DIVIDIDO {b} o RESTO da divisão é igual a : {resto}")
print("")
print("------------------------------POTENCIAÇÃO------------------------------")
print("")
print(f"O resultado da POTENCIACAO de {a} ELEVADO {b} é igual a : {potencia}")
print("")
print("-----------------------------------------------------------------------")
```

```python
# Operadores relacionais


# Entrada

a = int (input("digite um número: "))
b = int (input("digite um número: "))

# Processamento

# comparacao

maior = a > b
menor = a < b
maiorigual = a >= b
menorigual = a <= b

# Saída

print (maior)
print (menor)
print (maiorigual)
print (menorigual)
```

```python
# Fazer um programa para converter a temperatura
# F= (C x 9/5) + 32

# Entrada
celsius = float(input("Digite a temperatura em Celsius: "))

# Processamento
fahrenheit = (celsius * 9/5) + 32

# Saída
print(f"{celsius}°C é equivalente a {fahrenheit}°F")
```

```python
# Fazer um programa que comverte km em metros

# Entrada
quilometros = float(input("Digite a distância em quilômetros: "))

# Processamento
metros = quilometros * 1000

# Saída
print(f"{quilometros} km é equivalente a {metros} metros")
```

```python
# Fazer um programa que converte horas em minutos

# Entrada
horas = float(input("Digite a quantidade de horas: "))

# Processamento
minutos = horas * 60

# Saída
print(f"{horas} horas é equivalente a {minutos} minutos")
```
