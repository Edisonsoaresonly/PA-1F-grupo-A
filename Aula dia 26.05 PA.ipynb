```python
# Calculadora com 4 operações

# Menu

# Opção 1 - Somar
# Opção 2 - Subtrair
# Opção 3 - Multiplicar
# Opção 4 - Divisão
# Opção 5 - Sair

def add(x, y):
    return x + y

def subtract(x, y):
    return x - y

def multiply(x, y):
    return x * y

def divide(x, y):
    if y == 0:
        return "Erro! Divisão por zero!"
    return x / y

while True:
    print("\n--- Calculadora com 4 operações ---")
    print("Opção 1 - Somar")
    print("Opção 2 - Subtrair")
    print("Opção 3 - Multiplicar")
    print("Opção 4 - Divisão")
    print("Opção 5 - Sair")

    choice = input("Escolha a operação (1/2/3/4/5): ")

    if choice == '5':
        print("Saindo da calculadora. Até mais!")
        break

    if choice in ('1', '2', '3', '4'):
        try:
            num1 = float(input("Digite o primeiro número: "))
            num2 = float(input("Digite o segundo número: "))
        except ValueError:
            print("Entrada inválida. Por favor, digite números!")
            continue

        if choice == '1':
            print(f"{num1} + {num2} = {add(num1, num2)}")
        elif choice == '2':
            print(f"{num1} - {num2} = {subtract(num1, num2)}")
        elif choice == '3':
            print(f"{num1} * {num2} = {multiply(num1, num2)}")
        elif choice == '4':
            result = divide(num1, num2)
            if isinstance(result, str):
                print(result)
            else:
                print(f"{num1} / {num2} = {result}")
    else:
        print("Opção inválida. Por favor, escolha uma opção entre 1 e 5!")
```

```python
# Jogo Inteligente

# 1 - par / impar
# 2 - Positivo
# 3 - Maior de 02 números
# 4 - Sair

def check_even_odd(number):
    if number % 2 == 0:
        return "Par"
    else:
        return "Ímpar"

def check_positive(number):
    if number > 0:
        return "Positivo"
    elif number < 0:
        return "Negativo"
    else:
        return "Zero"

def find_greater(num1, num2):
    if num1 > num2:
        return f"O maior número é {num1}"
    elif num2 > num1:
        return f"O maior número é {num2}"
    else:
        return "Ambos os números são iguais"

while True:
    print("\n--- Jogo Inteligente ---")
    print("1 - Par / Ímpar")
    print("2 - Positivo / Negativo")
    print("3 - Maior de 02 números")
    print("4 - Sair")

    choice = input("Escolha uma opção (1/2/3/4): ")

    if choice == '4':
        print("Saindo do Jogo Inteligente. Até mais!")
        break

    if choice in ('1', '2', '3'):
        try:
            if choice == '1' or choice == '2':
                number = int(input("Digite um número: "))
                if choice == '1':
                    print(f"O número {number} é {check_even_odd(number)}")
                elif choice == '2':
                    print(f"O número {number} é {check_positive(number)}")
            elif choice == '3':
                num1 = float(input("Digite o primeiro número: "))
                num2 = float(input("Digite o segundo número: "))
                print(find_greater(num1, num2))
        except ValueError:
            print("Entrada inválida. Por favor, digite um número válido.")
            continue
    else:
        print("Opção inválida. Por favor, escolha uma opção entre 1 e 4.")
```

```python
# Escrever um programa que verifica a senha, o usuário terá 3 tentativas caso erre as 03 imprimir mensagem de "acesso negado e tente novamente mais tarde!" e encerrar o programa.

correct_password = "070722"
attempts = 0
max_attempts = 3

while attempts < max_attempts:
    user_input = input("Digite a senha: ")

    if user_input == correct_password:
        print("Acesso concedido! Bem-vindo!")
        break
    else:
        attempts += 1
        remaining_attempts = max_attempts - attempts
        if remaining_attempts > 0:
            print(f"Senha incorreta. Você tem mais {remaining_attempts} tentativa(s).")
        else:
            print("Senha incorreta. Você esgotou suas tentativas.")

if attempts == max_attempts:
    print("Acesso negado e tente novamente mais tarde!")
```
