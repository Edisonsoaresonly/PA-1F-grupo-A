```python
# escrever um programa que recebe as notas dos 04 bimestres e imprimi a média
# com o resultado de aprovado, recuperação, ou reprovado.

# Regras
 # Utilizar while;
 # nota igual ou maior que 7 aprovado;
 # nota igual ou maior que 5 recuperação;
 # nota igual que 5 reprovado.

nota = 0
bimestre = 1
resultado = 0

while bimestre <=4:
  nota = float(input(f"Digite a sua nota do {biestre}º bimestre: "))
  bimestre +=1
  resultado = resultado + nota

if (resultado /4) >=7:
  print(f"Sua média foi {resultado /4}, você está aprovado")
elif(resultado /4) >=5:
  print(f"Sua média foi {resultado /4}, você está de recuperação")
else;
  print(f"Sua média foi {resultado /4}, você está reprovado")
```

```python
# Projeto caixa eletrônico

# Você foi contratado para desenvolver um sistema simpes de caixa eletrônico.

# O sistema deverá funcionar através de um menu interativo utilizando "while"

# O programa deve iniciar com um saldo inicial de R$1000,00 e permitiu que o usuário escolha diferentes operações até decidir sair do sistema

# Ver saldo
# Depositar
# Sacar dinheiro
# Sair

# Regras:

# Quando depositar somar ao saldo anterior;
# Quando sacar subtrair do saldo anterior
# Mostrar o saldo para o usuário

saldo = 1000.00

while True:
    print("\n--- CAIXA ELETRÔNICO ---")
    print("1. Ver saldo")
    print("2. Depositar")
    print("3. Sacar dinheiro")
    print("4. Sair")

    try:
        opcao = int(input("Escolha uma opção: "))

        if opcao == 1:
            print(f"Seu saldo atual é: R${saldo:.2f}")
        elif opcao == 2:
            valor_deposito = float(input("Digite o valor a ser depositado: R$"))
            if valor_deposito > 0:
                saldo += valor_deposito
                print(f"Depósito de R${valor_deposito:.2f} realizado com sucesso.")
                print(f"Novo saldo: R${saldo:.2f}")
            else:
                print("Valor de depósito inválido.")
        elif opcao == 3:
            valor_saque = float(input("Digite o valor a ser sacado: R$"))
            if valor_saque > 0:
                if saldo >= valor_saque:
                    saldo -= valor_saque
                    print(f"Saque de R${valor_saque:.2f} realizado com sucesso.")
                    print(f"Novo saldo: R${saldo:.2f}")
                else:
                    print("Saldo insuficiente.")
            else:
                print("Valor de saque inválido.")
        elif opcao == 4:
            print("Obrigado por usar nosso caixa eletrônico. Volte sempre!")
            break
        else:
            print("Opção inválida. Por favor, escolha uma opção de 1 a 4.")
    except ValueError:
        print("Entrada inválida. Por favor, digite um número para a opção.")
```
