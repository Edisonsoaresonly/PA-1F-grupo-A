```python
# Escreva um número e informe ao usuário se ele é par ou ímpar

try:
    numero = int(input("Digite um número inteiro: "))

    if numero % 2 == 0:
        print(f"O número {numero} é PAR.")
    else:
        print(f"O número {numero} é ÍMPAR.")
except ValueError:
    print("Entrada inválida. Por favor, digite um número inteiro válido.")
```

```python
# Solicite dois números e mostre qual é o maior (ou se são iguais).

try:
    numero1 = float(input("Digite o primeiro número: "))
    numero2 = float(input("Digite o segundo número: "))

    if numero1 > numero2:
        print(f"O primeiro número ({numero1}) é maior que o segundo número ({numero2}).")
    elif numero2 > numero1:
        print(f"O segundo número ({numero2}) é maior que o primeiro número ({numero1}).")
    else:
        print(f"Os dois números são iguais ({numero1}).")
except ValueError:
    print("Entrada inválida. Por favor, digite números válidos.")
```

```python
# Peça a nota de um aluno e informe: o	Aprovado (nota ≥ 7) o	Recuperação (5 ≤ nota < 7) o	Reprovado (nota < 5)

# Entrada

nota = float(input("Digite a sua nota: "))

# Processamento / Saída

if nota >= 7:
  print("Aprovado")

elif nota >= 5 and nota <= 7:
  print("Recuperação")

else:
  print("Reprovado")
```

```python
# Peça um número e diga se ele é positivo, negativo ou zero.

# Entrada
try:
    numero = float(input("Digite um número: "))

# Processamento / Saída
    if numero > 0:
        print(f"O número {numero} é POSITIVO.")
    elif numero < 0:
        print(f"O número {numero} é NEGATIVO.")
    else:
        print(f"O número {numero} é ZERO.")
except ValueError:
    print("Entrada inválida. Por favor, digite um número válido.")
```

```python
# Peça dois números e uma operação (+, -, *, /). Use if/elif para calcular o
# resultado

try:
    # Entrada dos números
    numero1 = float(input("Digite o primeiro número: "))
    numero2 = float(input("Digite o segundo número: "))

    # Entrada da operação
    operacao = input("Digite a operação (+, -, *, /): ")

    # Processamento / Saída
    resultado = 0
    if operacao == '+':
        resultado = numero1 + numero2
        print(f"Resultado: {numero1} + {numero2} = {resultado}")
    elif operacao == '-':
        resultado = numero1 - numero2
        print(f"Resultado: {numero1} - {numero2} = {resultado}")
    elif operacao == '*':
        resultado = numero1 * numero2
        print(f"Resultado: {numero1} * {numero2} = {resultado}")
    elif operacao == '/':
        if numero2 != 0:
            resultado = numero1 / numero2
            print(f"Resultado: {numero1} / {numero2} = {resultado}")
        else:
            print("Erro: Divisão por zero não é permitida.")
    else:
        print("Operação inválida. Por favor, use +, -, * ou /.")

except ValueError:
    print("Entrada inválida. Por favor, digite números válidos.")
```

```python
# Peça um número de 1 a 7 e mostre o dia correspondente (ex: 1 = domingo).

try:
    numero_dia = int(input("Digite um número de 1 a 7 para ver o dia da semana: "))

    if numero_dia == 1:
        print("1 = Domingo")
    elif numero_dia == 2:
        print("2 = Segunda-feira")
    elif numero_dia == 3:
        print("3 = Terça-feira")
    elif numero_dia == 4:
        print("4 = Quarta-feira")
    elif numero_dia == 5:
        print("5 = Quinta-feira")
    elif numero_dia == 6:
        print("6 = Sexta-feira")
    elif numero_dia == 7:
        print("7 = Sábado")
    else:
        print("Número inválido. Por favor, digite um número entre 1 e 7.")

except ValueError:
    print("Entrada inválida. Por favor, digite um número inteiro válido.")
```

```python
# Solicite um ano e informe se ele é bissexto ou não.

# Entrada

try:
    ano = int(input("Digite um ano: "))

    # Processamento / Saída
    if (ano % 4 == 0 and ano % 100 != 0) or (ano % 400 == 0):
        print(f"O ano {ano} é bissexto.")
    else:
        print(f"O ano {ano} NÃO é bissexto.")
except ValueError:
    print("Entrada inválida. Por favor, digite um número inteiro válido para o ano.")
```

```python
# Peça o valor de uma compra: ≥ 100: 10% de desconto ≥ 50 e < 100: 5% de desconto < 50: sem desconto

# Entrada
try:
    valor_compra = float(input("Digite o valor da compra: "))

# Processamento / Saída

    if valor_compra >= 100:
        desconto = valor_compra * 0.10
        valor_final = valor_compra - desconto
        print(f"Valor da compra: R${valor_compra:.2f}")
        print(f"Desconto de 10%: R${desconto:.2f}")
        print(f"Valor final com desconto: R${valor_final:.2f}")
    elif valor_compra >= 50 and valor_compra < 100:
        desconto = valor_compra * 0.05
        valor_final = valor_compra - desconto
        print(f"Valor da compra: R${valor_compra:.2f}")
        print(f"Desconto de 5%: R${desconto:.2f}")
        print(f"Valor final com desconto: R${valor_final:.2f}")
    else:
        valor_final = valor_compra
        print(f"Valor da compra: R${valor_compra:.2f}")
        print("Sem desconto.")
        print(f"Valor final: R${valor_final:.2f}")

except ValueError:
    print("Entrada inválida. Por favor, digite um número válido para o valor da compra.")
```
