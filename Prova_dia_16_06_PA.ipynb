```python
# Solicite ao usuário um número inteiro positivo.

# Utilizando um laço de repetição, calcule e exiba a soma de
# todos os números de 1 até o número informado.

# Exemplo:

# Digite um número: 5

# Resultado: 15

# Pois:

# 1 + 2 + 3 + 4 + 5 = 15

numero = int(input("Digite um número inteiro positivo: "))

soma = 0
for i in range(1, numero + 1):
    soma += i

print(f"Resultado: {soma}")
```

```python
# Solicite ao usuário um número inteiro positivo e calcule
# seu fatorial utilizando um laço de repetição.

# Exemplo:

# Digite um número: 5

# Fatorial = 120

# Pois:

# 5! = 5 × 4 × 3 × 2 × 1

# Observação:
# O programa deve validar se o número informado é positivo.

numero = int(input("Digite um número inteiro positivo: "))

if numero < 0:
    print("Erro: O número deve ser positivo.")
elif numero == 0:
    print("Fatorial = 1")
else:
    fatorial = 1
    for i in range(1, numero + 1):
        fatorial *= i
    print(f"Fatorial = {fatorial}")
```

```python
# Um caixa eletrônico possui notas de:
# R$ 100
# R$ 50
# R$ 20
# R$ 10
# R$ 5
# R$ 2
# R$ 1

# Solicite ao usuário o valor do saque e informe a menor
# quantidade possível de notas para realizar o pagamento.

# Exemplo:

# Digite o valor do saque: 287

# Saída esperada:

# 2 nota(s) de R$100
# 1 nota(s) de R$50
# 1 nota(s) de R$20
# 1 nota(s) de R$10
# 1 nota(s) de R$5
# 1 nota(s) de R$2

# Requisitos:
#- Utilizar laços de repetição.
#- Não utilizar listas.
#- Exibir apenas as notas utilizadas.

valor_saque = int(input("Digite o valor do saque: "))
valor_restante = valor_saque

print("\nSaída esperada:\n")

# Notas de R$100
quantidade_100 = 0
while valor_restante >= 100:
    quantidade_100 += 1
    valor_restante -= 100
if quantidade_100 > 0:
    print(f"{quantidade_100} nota(s) de R$100")

# Notas de R$50
quantidade_50 = 0
while valor_restante >= 50:
    quantidade_50 += 1
    valor_restante -= 50
if quantidade_50 > 0:
    print(f"{quantidade_50} nota(s) de R$50")

# Notas de R$20
quantidade_20 = 0
while valor_restante >= 20:
    quantidade_20 += 1
    valor_restante -= 20
if quantidade_20 > 0:
    print(f"{quantidade_20} nota(s) de R$20")

# Notas de R$10
quantidade_10 = 0
while valor_restante >= 10:
    quantidade_10 += 1
    valor_restante -= 10
if quantidade_10 > 0:
    print(f"{quantidade_10} nota(s) de R$10")

# Notas de R$5
quantidade_5 = 0
while valor_restante >= 5:
    quantidade_5 += 1
    valor_restante -= 5
if quantidade_5 > 0:
    print(f"{quantidade_5} nota(s) de R$5")

# Notas de R$2
quantidade_2 = 0
while valor_restante >= 2:
    quantidade_2 += 1
    valor_restante -= 2
if quantidade_2 > 0:
    print(f"{quantidade_2} nota(s) de R$2")

# Notas de R$1
quantidade_1 = 0
while valor_restante >= 1:
    quantidade_1 += 1
    valor_restante -= 1
if quantidade_1 > 0:
    print(f"{quantidade_1} nota(s) de R$1")
```
