```python
# Estrutura de decisões

# Escrever um programa que recebe o nome e a idade do usuário e verifica se o usuário é criança, adolescente e adulto

# >= 12 anos é criança
# =< 17 anos é adolescente
# Adulto

# Entrada

nome = str(input("digite o seu nome: "))
idade = int(input("digite sua idade: "))

# Processamento / Saída

if idade <= 12:
 print(f"{nome}, você tem {idade} anos de idade, você é criança.")

elif idade <= 17:
  print(f"{nome}, você tem {idade} anos de idade, você é adolescente.")

else:
  print(f"{nome}, você tem {idade} anos de idade, você é adulto.")
```

```python
# Em um jogo, um personagem possui uma variavel chamada energia. O programa deve: Se a energia for maior que 70 - Forte, se a energia estiver entre 30 e 70 - Cansado, se a energia estiver menor que 30 - Fraco.

# Entrada

Energia = int(input("digite a energia do seu personagem: "))

# Processamento

if Energia <= 30:
  print(f"Seu personagem tem energia igual a {Energia}. Ele está fraco.")

elif Energia <= 70:
  print(f"Seu personagem tem energia igual a {Energia}. Ele está cansado.")

else:
   print(f"Seu personagem tem energia igual a {Energia}. Ele está forte.")
```
