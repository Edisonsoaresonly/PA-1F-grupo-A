```python
numero = int(input("Digite um número: "))

if numero > 0:
  for i in range(1, numero + 1):
    print(i)

else:
  print("Por favor, digite um número inteiro positivo.")
```

```python
numero = int(input("Digite um número para a contagem regressiva: "))

if numero >= 0:
  for i in range(numero, -1, -1):
    print(i)

else:
  print("Por favor, digite um número inteiro não negativo.")
```

```python
numero = int(input("Digite um número: "))

print(f"Números pares até {numero}:")
if numero > 0:
  for i in range(1, numero + 1):
    if i % 2 == 0:
      print(i)

else:
  print("Por favor, digite um número inteiro positivo.")
```

```python
numero = int(input("Digite um número: "))

print(f"Números ímpares até {numero}:")
if numero > 0:
  for i in range(1, numero + 1):
    if i % 2 != 0:
      print(i)

else:
  print("Por favor, digite um número inteiro positivo.")
```

```python
numero = int(input("Digite um número: "))

if numero > 0:
  soma = 0
  for i in range(1, numero + 1):
    soma += i
  print(f"A soma dos números até {numero} é: {soma}")

else:
  print("Por favor, digite um número inteiro positivo.")
```

```python
palavra = input("Digite uma palavra: ")

numero_letras = len(palavra)
print(f"A palavra '{palavra}' tem {numero_letras} letras.")
```

```python
palavra = input("Digite uma palavra: ")
vogais = "aeiouAEIOU"
contador_vogais = 0

for letra in palavra:
  if letra in vogais:
    contador_vogais += 1

print(f"A palavra '{palavra}' tem {contador_vogais} vogais.")
```

```python
palavra = input("Digite uma palavra: ")
palavra_invertida = palavra[::-1]
print(f"A palavra invertida é: {palavra_invertida}")
```

```python
print("Digite quatro números:")

numero1 = int(input("Primeiro número: "))
numero2 = int(input("Segundo número: "))
numero3 = int(input("Terceiro número: "))
numero4 = int(input("Quarto número: "))

maior = numero1

if numero2 > maior:
  maior = numero2

if numero3 > maior:
  maior = numero3

if numero4 > maior:
  maior = numero4

print(f"O maior número digitado é: {maior}")
```
