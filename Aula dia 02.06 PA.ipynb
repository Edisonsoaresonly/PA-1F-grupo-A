```python
# Fazer um jogo de adivinhação

# O usuário irá digitar um número de 0 a 10 para adivinhar o número secreto
# O usuário terá 03 tentativas para acertar o número
# O programa deverá imprimir uma mensagem se o palpite do usuário é maior ou menor que o número secreto
# O programa deverá imprimir uma mensagem de quente, morno ou frio

# Regras

# Se a diferença do palpite para o número secreto for 3 - frio, 2 - morno e 1 quente

import random

numero_secreto = random.randint(0, 10)
chances = 3

print("Bem-vindo ao jogo de adivinhação!")
print("Estou pensando em um número entre 0 e 10.")

for i in range(chances):
    try:
        guess = int(input(f"Tentativa {i + 1}/{chances}. Qual é o seu palpite? "))

        if not (0 <= guess <= 10):
            print("Por favor, digite um número entre 0 e 10.")
            continue

        if guess == numero_secreto:
            print(f"Parabéns! Você adivinhou o número secreto ({numero_secreto}) em {i + 1} tentativas!")
            break
        else:
            if guess < numero_secreto:
                print("Seu palpite é muito baixo.")
            else:
                print("Seu palpite é muito alto.")

            difference = abs(numero_secreto - guess)
            if difference == 1:
                print("Está quente.")
            elif difference == 2:
                print("Está morno.")
            elif difference >= 3:
                print("Está frio.")
    except ValueError:
        print("Entrada inválida. Por favor, digite um número inteiro.")

else:
    print(f"Suas tentativas acabaram! O número secreto era {numero_secreto}.")
```
