Calcular a área do retangulo

```python
# Entrada
comprimento = float(input("Digite o comprimento do retângulo:"))
altura = float(input("Digite a altura do retângulo:"))

# Processamento

área = comprimento * altura


# Saída

print(f"O total da área do retângulo é:{area}")
```

Fazer um programa que calcula o dobro de um número

Entrada: Um número

Processamento: Calcular o dobro

Saída: Mostrar o resultado

```python
# Entrada

Numero = int (input("Digite um número:"))

# Processamento

Dobro = Numero * 2

# Saída

print(f"O dobro do numero {Numero} é igual a {Dobro}")
```
