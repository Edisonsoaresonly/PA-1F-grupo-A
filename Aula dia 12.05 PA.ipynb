```python
# 10 - Pirâmide Numérica Simétrica

# Faça um programa que receba um número N e imprima uma pirâmide perfeitamente centralizada.

def piramide_numerica_simetrica(n):
    if n <= 0:
        print("Por favor, insira um número inteiro positivo.")
        return

    # Calcular a largura total da base para centralização
    # Corrigido: Converta ambos os ranges para listas antes de concatenar
    max_width = len(' '.join(map(str, list(range(1, n + 1)) + list(range(n - 1, 0, -1))))) # Example for n=3: 1 2 3 2 1

    for i in range(1, n + 1):
        # Parte crescente
        row_elements = list(range(1, i + 1))
        # Parte decrescente (excluindo o último número da parte crescente)
        if i > 1:
            row_elements.extend(range(i - 1, 0, -1))

        row_str = ' '.join(map(str, row_elements))
        print(row_str.center(max_width))

# Exemplo de uso:
# Você pode alterar o valor de N aqui para testar diferentes tamanhos de pirâmide
N = 5 # Defina o número N para a pirâmide
piramide_numerica_simetrica(N)
```
